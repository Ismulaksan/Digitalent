```python
import pandas as pd
```

```python
data_impor = pd.read_csv ('nilai-impor-negara-asal-2005-2013.csv', delimiter = ',', encoding = "ISO-8859-1")
```

```python
data_impor[['tahun','nama_negara','region','nilai_impor']]
```

```python
data_impor.groupby(['region','nama_negara','nilai_impor']).sum()
```

```python
ax = data_impor.groupby('region').sum()['nilai_impor']
ax = ax.plot(kind='bar', title ="Total impor tiap region", legend=True, fontsize=12)
ax.set_xlabel("Negara", fontsize=12)
ax.set_ylabel("Nilai impor", fontsize=12)
```

dari hasil visualisasi diatas dapat disimpulkan bahwa impor negara indonesia didominasi oleh negara-negara yang berada pada region Asia Timur, diikuti region ASEAN, Uni Eropa, Amerika Utara dan lainnya

```python
data_impor.sort_values(by=['nilai_impor'],ascending=False)[:20]
```

```python
data_impor.groupby('tahun').mean()
```

```python
data_impor.groupby('tahun').sum()
```

```python
ax = data_impor.groupby('tahun').sum()['nilai_impor']
ax = ax.plot(kind='bar', title ="Nilai impor per tahun", legend=True, fontsize=12)
ax.set_xlabel("tahun", fontsize=12)
ax.set_ylabel("Nilai impor", fontsize=12)
```

impor indonesia cenderung mmeningkat dari tahun ke tahun

```python
data_impor.groupby('nama_negara').sum()['nilai_impor'].sort_values(ascending=False)[:10]
```

```python
ax = data_impor.groupby('nama_negara').sum()['nilai_impor']
ax = ax.plot(kind='bar', title ="Nilai impor tiap negara", legend=True, fontsize=8)
ax.set_xlabel("nama negara", fontsize=12)
ax.set_ylabel("Nilai impor", fontsize=12)
```

dari visualisasi diatas menggambarkan bahwa impor indonesia terbesar yaitu dari negara Singapura, RRC, Jepang USA
